__Important Libraries__

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from bs4 import BeautifulSoup
import seaborn as sns
import nltk
from wordcloud import WordCloud
from nltk.corpus import stopwords
from nltk.util import ngrams
from nltk.probability import FreqDist
from nltk.tokenize import word_tokenize, sent_tokenize
from bs4 import BeautifulSoup


```

__Null Values and Duplicates__

```python
df=pd.read_csv('IMDB Dataset.csv').sample(1000)
len(df)
```

```python
# Check for null values
null_values = df.isnull().sum()

# Check for duplicates
duplicates = df.duplicated().sum()

null_values, duplicates
```

__Remove HTML__

```python
def clean_html(raw_html):
    clean_text = BeautifulSoup(raw_html, "html.parser").get_text()
    return clean_text
df['review'] = df['review'].apply(clean_html)
```

__Word Count (with and without stopwords)__

```python
stop_words = set(stopwords.words('english'))

# Tokenize the reviews
df['word_count'] = df['review'].apply(lambda x: len(word_tokenize(x)))
df['word_count_no_stopwords'] = df['review'].apply(lambda x: len([word for word in word_tokenize(x) if word.lower() not in stop_words]))

# Most common and rare words
all_words = ' '.join(df['review']).lower()
all_words = word_tokenize(all_words)
word_freq = FreqDist(all_words)
common_words = word_freq.most_common(10)
rare_words = [word for word, freq in word_freq.items() if freq == 1]

df[['review', 'word_count', 'word_count_no_stopwords']], common_words, rare_words
```

__Distribution of Sentence Lengths__

```python
# Plotting the distribution of sentence lengths
plt.figure(figsize=(10, 6))
sns.histplot(df['word_count'], kde=True)
plt.title('Distribution of Sentence Lengths')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.show()
```

__N-gram Pairs__

```python
bigrams = list(ngrams(all_words, 2))
trigrams = list(ngrams(all_words, 3))

bigram_freq = FreqDist(bigrams)
trigram_freq = FreqDist(trigrams)

common_bigrams = bigram_freq.most_common(10)
common_trigrams = trigram_freq.most_common(10)

common_bigrams, common_trigrams
```

__Sentence Types__

```python
# Classifying sentence types
df['sentence_type'] = df['review'].apply(lambda x: 'exclamatory' if '!' in x else 'interrogative' if '?' in x else 'declarative')

df['sentence_type'].value_counts()
```

__POS Tagging__

```python
# POS Tagging
df['pos_tags'] = df['review'].apply(lambda x: nltk.pos_tag(word_tokenize(x)))

df[['review', 'pos_tags']]
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

__Named Entity Recognition (NER)__

```python
# NER
df['named_entities'] = df['review'].apply(lambda x: nltk.ne_chunk(nltk.pos_tag(word_tokenize(x))))

df[['review', 'named_entities']]
```

__Count of Vowels__

```python
vowels = 'aeiou'

df['vowel_count'] = df['review'].apply(lambda x: sum([1 for char in x.lower() if char in vowels]))

df[['review', 'vowel_count']]
```

__Semantic Analysis__

```python
# Sentiment scores using TextBlob (or similar library)
from textblob import TextBlob

df['polarity'] = df['review'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['subjectivity'] = df['review'].apply(lambda x: TextBlob(x).sentiment.subjectivity)

df[['review', 'polarity', 'subjectivity']]
```

__Syntactic Analysis__

```python
# Dependency parsing (using spaCy)
import spacy

nlp = spacy.load('en_core_web_sm')

df['dependency_parsing'] = df['review'].apply(lambda x: [(token.text, token.dep_, token.head.text) for token in nlp(x)])

df[['review', 'dependency_parsing']]
```

__Word Cloud__

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(df['review']))

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Reviews')
plt.show()
```

__Topic Modeling__

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Vectorizing the text data
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['review'])

# Fitting LDA model
lda = LatentDirichletAllocation(n_components=2, random_state=42)
lda.fit(X)

# Displaying the topics
def display_topics(model, feature_names, no_top_words):
    topics = []
    for topic_idx, topic in enumerate(model.components_):
        topics.append([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]])
    return topics

no_top_words = 5
feature_names = vectorizer.get_feature_names_out()
topics = display_topics(lda, feature_names, no_top_words)

topics
```
